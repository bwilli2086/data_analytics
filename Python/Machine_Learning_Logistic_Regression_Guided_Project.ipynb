# Logistic Regression

Now it's time to do a train test split, and train the model.

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
ds=ad_data
y=ds['clicked_on_ad']
X=ds[['daily_time_spent_on_site','age','area_income','daily_internet_usage','male']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

** Train and fit a logistic regression model on the training set.**

```python
logmodel=LogisticRegression
```

```python
logmodel.fit(X_train,y_train)
```

```python
LogisticRegression(C=1.0,class_weight=None,dual=False,fit_intercept=True,intercept_scaling=1,l1_ratio=None,max_iter=100,multi_class='auto',n_jobs=None,penalty='12',random_state=None,solver='lbfgs',tol=0.0001,verbose=0,warm_start=False)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
y_test_pred=logmodel.predict()
```

** Create a classification report for the model.**

```python
print(classification_report(y_test,y_test_pred))
print('\n')
print(confusion_matrix(y_test,y_test_pred))
```
